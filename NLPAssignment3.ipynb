```python
import re
from collections import defaultdict, Counter

class NGramModel:
    def __init__(self, n):
        self.n = n
        self.ngrams = defaultdict(Counter)
        self.vocab = set()

    def preprocess(self, text):
        text = text.lower()
        text = re.sub(r'[^a-z\s]', '', text)
        tokens = text.split()
        return tokens

    def train(self, text):
        tokens = self.preprocess(text)
        self.vocab.update(tokens)

        for i in range(len(tokens) - self.n + 1):
            context = tuple(tokens[i:i+self.n-1])
            next_word = tokens[i+self.n-1]
            self.ngrams[context][next_word] += 1

    def predict_next(self, text):
        tokens = self.preprocess(text)
        context = tuple(tokens[-(self.n-1):])

        if context not in self.ngrams:
            return "No prediction available"

        next_words = self.ngrams[context]
        most_probable = next_words.most_common(1)[0][0]
        return most_probable

    def get_probabilities(self, text):
        tokens = self.preprocess(text)
        context = tuple(tokens[-(self.n-1):])

        if context not in self.ngrams:
            return {}

        next_words = self.ngrams[context]
        total = sum(next_words.values())

        probs = {word: count/total for word, count in next_words.items()}
        return probs
```

```python
corpus = """
machine learning is fun
machine learning is powerful
machine learning models learn patterns
deep learning is part of machine learning
"""

model = NGramModel(3)   # Trigram model
model.train(corpus)

print("Prediction:", model.predict_next("machine learning"))
print("Probabilities:", model.get_probabilities("machine learning"))
```
