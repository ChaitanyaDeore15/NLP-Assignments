```python
def extract_noun_phrases(paragraph):
    NP_List = []
    sentences = sent_tokenize(paragraph)

    for S in sentences:
        words = word_tokenize(S)
        tagged_words = pos_tag(words)
        grammar = r"NP: {<DT>?<JJ>*<NN|NNS|NNP|NNPS>+}"
        cp = RegexpParser(grammar)
        parse_tree = cp.parse(tagged_words)
        for subtree in parse_tree.subtrees(filter=lambda t: t.label() == 'NP'):
            np = " ".join(word for word, tag in subtree.leaves())
            NP_List.append(np)

    return NP_List
```

```python
paragraph = """
Artificial Intelligence is transforming the modern world.
Machine learning models are widely used in healthcare, finance, and education systems.
The rapid growth of technology is creating new opportunities for businesses.
"""
```

```python
noun_phrases = extract_noun_phrases(paragraph)
```

```python
print("Extracted Noun Phrases:\n")

for i, np in enumerate(noun_phrases, 1):
    print(f"{i}. {np}")
```
